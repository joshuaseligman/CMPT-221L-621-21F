## What's the difference between Null and undefined?

Null is where you create a variable and purposely define it as nothing.

On the other side, undefined is where a variable is declared but not defined yet. In other words, since the computer does not know what the variable is supposed to store, it says it is undefined until it is given a value.

## List the 3 Variable Types and Use Them!

```javascript
// The 3 variable types are var, let, and const

var myVar = 5;
myVar = 10;
var myVar = 25;

let myLet = 6;
myLet = 1;

const MY_CONST = 500;

console.log(`myVar is ${myVar}. myLet is ${myLet}. MY_CONST is ${MY_CONST}`);
```

## Use an If/Else If/Else Statement and a Switch Case statement

```javascript
if (myVar < 5) {
    console.log('Tiny');
} else if (myVar < 10) {
    console.log('Medium');
} else {
    console.log('Big');
}

switch (myLet) {
    case 0:
        console.log('Nothing');
        break;
    case 1:
        console.log('One thing');
        break;
    case 2:
        console.log('Multiple things');
        break;
    default:
        console.log('What is this?');
        break;
}
```

## Use a For and a While Loop

```javascript
let sum = 0;
for (let i = 0; i < 10; i++) {
    sum += i;
    console.log(sum);
}

console.log();

let x = 100;
while (x > 0.5) {
    console.log(x);
    x /= 2;
}
```
